```python
import re
import matplotlib.pyplot as plt
```

## Identity activation, 1 hidden layer

```python
train_losses, train_accuracies, val_losses, val_accuracies = [], [], [], []

with open('training_model_Identity_1_layer.log', 'r') as file:
    lines = file.readlines()
    for line in lines:
        match = re.search(r'Train Loss: ([\d.]+), Train Acc: ([\d.]+), Val Loss: ([\d.]+), Val Acc: ([\d.]+)', line)
        if match:
            train_losses.append(float(match.group(1)))
            train_accuracies.append(float(match.group(2)))
            val_losses.append(float(match.group(3)))
            val_accuracies.append(float(match.group(4)))


epochs = range(len(train_losses))
plt.figure(figsize=(10, 5))

# Wykres funkcji straty
plt.subplot(1, 2, 1)
plt.plot(epochs, train_losses, label='Train')
plt.plot(epochs, val_losses, label='Validation')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Wykres dokładności
plt.subplot(1, 2, 2)
plt.plot(epochs, train_accuracies, label='Train')
plt.plot(epochs, val_accuracies, label='Validation')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
